```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
```

```python
df = pd.read_csv('games.csv')
```

```python
df = df[['turns','winner']]
df['value'] = 1
df.sort_values('turns').head(20)
```

```python
df = df.groupby(['turns','winner'])
df = df.count()
df.head(20)
```

```python
df = df.unstack()
df = df['value']
```

```python
composit_games = list()
for i in range(5,len(df),5):
    composit_games.append(list(df.loc[i-5:i].sum()))
composit_games
```

```python
df2 = pd.DataFrame(composit_games)
```

```python
data_perc = df2.divide(df2.sum(axis=1), axis=0)
data_perc.head()
# Converted as a percentage
```

```python
data_perc.head()
```

```python
mpl.rcParams['figure.figsize'] = (50,50)
```

```python
data_perc.plot(kind='bar', stacked=True)
# Propasal: group the x axis (the turns/indices of df) by 10s. 1-10 turns, 11-20 turns, etc...)
```
