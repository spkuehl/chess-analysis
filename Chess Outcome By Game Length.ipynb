```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
```

```python
df = pd.read_csv('games.csv')
```

```python
df = df[['turns','winner']]
df['value'] = 1
df.head()
```

```python
df = df.groupby(['turns','winner'])
df = df.count()
df.head()
```

```python
df = df.unstack()
df = df['value']
df.head()
```

```python
data_perc = df.divide(df.sum(axis=1), axis=0)
data_perc.head()
# Converted as a percentage
```

```python
mpl.rcParams['figure.figsize'] = (50,50)
```

```python
data_perc.plot(kind='bar', stacked=True)
# Propasal: group the x axis (the turns/indices of df) by 10s. 1-10 turns, 11-20 turns, etc...)
```
